### Carregando os datasets

## Dataframe de todos os jogos

```python
import re

colunas = ['ID','data','hora','visitante_Placar','mandante_Placar','arena','mandante','visitante','rodata']
df_limpo = df[colunas].copy()
df_limpo = df_limpo.rename(columns={'visitante_Placar': 'n_placar_visitante','mandante_Placar': 'n_placar_casa','arena': 'c_nome_estadio','mandante': 'c_time_casa','visitante': 'c_time_visitante','rodata':'n_rodada'})

df_limpo['data'] = pd.to_datetime(df_limpo['data'], format='%d/%m/%Y')
df_limpo = df_limpo[df_limpo['data'].dt.year >= 2014] # só temos infomações dos gols a partir desse ano
df_limpo ['c_nome_estadio'] = df_limpo ['c_nome_estadio'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
df_limpo['d_ano_campeonato'] = df_limpo['data'].dt.year
df_limpo['c_nome_campeonato'] = "Brasileirão " + df_limpo['d_ano_campeonato'].astype(str)
df_limpo.index = range(1, len(df_limpo) + 1)


display(df_limpo)
df_limpo.drop(columns=['ID']).to_csv("inserts/jogos_brasileirão.csv", index=False)
```

### Código para inserção de todos os estádios

```python
estadios = []
for estadio in df_limpo['c_nome_estadio'].unique():
    row = f"INSERT IGNORE INTO Estadio (c_nome_estadio, c_cidade_estadio, n_capacidade) VALUES ('{estadio}', NULL, 45000);"
    estadios.append(row)

# 4. Salvar o resultado em um arquivo de texto
DIR = 'inserts/insert_estadios.txt'
try:
    with open(DIR, 'w', encoding='utf-8') as f:
        f.write('\n'.join(estadios))
    print(f"Arquivo '{DIR}' criado com sucesso!")
except Exception as e:
    print(f"Ocorreu um erro ao salvar o arquivo: {e}")
```

## Dataframe dos jogadores

```python
# Criar as colunas ausentes corretamente
df_jogador_gols = df_gols[['atleta', 'clube']].copy()
df_jogador_gols['num_camisa'] = pd.NA
df_jogador_gols['posicao'] = pd.NA

df_jogador_cartoes = df_cartoes[['atleta', 'clube', 'num_camisa', 'posicao']].copy()

# Unir os dois DataFrames
df_jogadores = pd.concat([df_jogador_gols, df_jogador_cartoes], ignore_index=True)

# Remover duplicados, mas mantendo uma linha por jogador por clube
df_jogadores_unicos = df_jogadores.drop_duplicates(subset=['atleta', 'clube'], keep='first').reset_index(drop=True)

# Exibir tudo
display(df_jogadores_unicos)
```

```python
def contar_nao_nulos(row):
    # Conta quantos campos não são NaN na linha (excluindo atleta e clube)
    return row.notna().sum()

# Contar os não nulos (exceto atleta e clube)
df_jogadores['completude'] = df_jogadores[['num_camisa', 'posicao']].notna().sum(axis=1)

# Ordenar para que as linhas mais completas fiquem primeiro
df_jogadores_ordenado = df_jogadores.sort_values(by='completude', ascending=False)

# Agora fazer o drop_duplicates por (atleta, clube), ficando com a linha mais completa
df_jogadores_unicos = df_jogadores_ordenado.drop_duplicates(subset=['atleta', 'clube'], keep='first').reset_index(drop=True)

# Remover a coluna auxiliar de completude
df_jogadores_unicos = df_jogadores_unicos.drop(columns='completude')

# Exibir
display(df_jogadores_unicos)
```

```python
import numpy as np
df_jogadores_unicos['posicao'] = df_jogadores_unicos['posicao'].str.strip()
df_jogadores_unicos['posicao'] = df_jogadores_unicos['posicao'].str.upper()

df_jogadores_unicos['atleta'] = df_jogadores_unicos['atleta'].str.strip()
df_jogadores_unicos['atleta'] = df_jogadores_unicos['atleta'].str.upper()

df_jogadores_unicos['clube'] = df_jogadores_unicos['clube'].str.strip()

df_jogadores_unicos['num_camisa'] = df_jogadores_unicos['num_camisa'].astype('Int64')

df_jogadores_unicos['atleta'] = df_jogadores_unicos['atleta'].astype(str)
df_jogadores_unicos['Pnome'] = df_jogadores_unicos['atleta'].str.split().str[0]
df_jogadores_unicos['Unome'] = df_jogadores_unicos['atleta'].str.split().str[-1]
condicao = df_jogadores_unicos['Pnome'] == df_jogadores_unicos['Unome']
df_jogadores_unicos.loc[condicao, 'Unome'] = np.nan
df_jogadores_unicos = df_jogadores_unicos.drop('atleta', axis=1)
df_jogadores_unicos['d_data_nascimento'] = pd.NA

colunas = ['Pnome','Unome','num_camisa','d_data_nascimento','posicao','clube']
df_jogadores_unicos = df_jogadores_unicos[colunas].copy()
df_jogadores_unicos = df_jogadores_unicos.rename(columns={'Pnome':'c_Pnome_jogador','Unome':'c_Unome_jogador','num_camisa': 'n_camisa','posicao': 'c_posicao','clube': 'c_nome_time'})


display(df_jogadores_unicos)

df_jogadores_unicos.to_csv("inserts/jogadores_brasileirao.csv", index = False)
```

## Dataframe de gols

```python
def converter_minuto(minuto_str):
    if isinstance(minuto_str, str) and '+' in minuto_str:
        partes = minuto_str.replace('+', ' ').split()
        return sum(map(int, partes))
    try:
        return int(minuto_str)
    except:
        return None
```

```python
#ARRUMANDO ID JOGADOR
df_gols['atleta'] = df_gols['atleta'].str.upper()
df_gols['atleta'] = df_gols['atleta'].astype(str)
df_gols['Pnome'] = df_gols['atleta'].str.split().str[0]
df_gols['Unome'] = df_gols['atleta'].str.split().str[-1]
condicao = df_gols['Pnome'] == df_gols['Unome']
df_gols.loc[condicao, 'Unome'] = np.nan

#ARRUMANDO MINUTO JOGO
df_gols['minuto'] = df_gols['minuto'].apply(converter_minuto)

def buscar_indice_jogador(row):
    pnome = row['Pnome']
    unome = row['Unome']
    clube = row['clube']

    cond = (df_jogadores_unicos['c_Pnome_jogador'] == pnome)
    
    # Verificar também o sobrenome (c_Unome_jogador), mas só se o sobrenome não for NaN
    if pd.notna(unome):
        cond = cond & (df_jogadores_unicos['c_Unome_jogador'] == unome)
    
    cond = cond & (df_jogadores_unicos['c_nome_time'] == clube)

    resultado = df_jogadores_unicos[cond]
    
    if not resultado.empty:
        return resultado.index[0]+1  # Pega o primeiro índice correspondente
    else:
        return np.nan  # Caso não encontre

# Criando a nova coluna com o índice
df_gols['id_jogador'] = df_gols.apply(buscar_indice_jogador, axis=1)
display(df_gols)
```

```python
#ARRUMANDO ID DO JOGO
#df_gols_2023 = df_gols[df_gols['partida_id'].isin(df_limpo['ID'])].copy()
id_para_index = df_limpo.reset_index().set_index('ID')['index']
df_gols['id_jogo'] = df_gols['partida_id'].map(id_para_index)

#ELIMINANDO COLUNAS DESNECESSARIAS
df_gols = df_gols.drop(['atleta','rodata','clube','tipo_de_gol','Pnome','Unome','partida_id'], axis=1)
df_gols = df_gols.rename(columns={'minuto': 'n_minuto_gol'})

#GUARDANDO GOLS EM CSV
df_gols.to_csv("inserts/gols_brasileirao.csv",index=False)
display(df_gols)
```

```python
colunas_chave = ['id_jogo', 'id_jogador', 'n_minuto_gol']

linhas_duplicadas = df_gols[df_gols.duplicated(subset=colunas_chave, keep=False)]

# 3. Verifique se foram encontradas duplicatas e mostre o resultado.
if not linhas_duplicadas.empty:
    print("--- ATENÇÃO: Linhas Duplicadas Encontradas no DataFrame ---")
    
    # Ordenar os resultados pelas colunas da chave ajuda a agrupar
    # visualmente as duplicatas, tornando a análise mais fácil.
    print(linhas_duplicadas.sort_values(by=colunas_chave))
else:
    print("--- Nenhuma linha duplicada foi encontrada dentro do próprio DataFrame. ---")
    print("\nSe você ainda recebe o erro do banco de dados, a causa é outra:")
    print("Uma ou mais linhas do seu DataFrame já existem na tabela 'Gol' do banco,")
    print("provavelmente de uma inserção anterior.")
```

```python
df_cartoes['posicao'] = df_cartoes['posicao'].str.strip()
df_cartoes['posicao'] = df_cartoes['posicao'].str.upper()

df_cartoes['atleta'] = df_cartoes['atleta'].str.upper()
df_cartoes['atleta'] = df_cartoes['atleta'].astype(str)
df_cartoes['Pnome'] = df_cartoes['atleta'].str.split().str[0]
df_cartoes['Unome'] = df_cartoes['atleta'].str.split().str[-1]
condicao = df_cartoes['Pnome'] == df_cartoes['Unome']
df_cartoes.loc[condicao, 'Unome'] = np.nan
df_cartoes = df_cartoes.drop('atleta', axis=1)
df_cartoes['minuto'] = df_cartoes['minuto'].apply(converter_minuto)

df_cartoes['id_jogador'] = df_cartoes.apply(buscar_indice_jogador, axis=1)

#ARRUMANDO ID DO JOGO
#df_cartoes_2023 = df_cartoes[df_cartoes['partida_id'].isin(df_limpo['ID'])].copy()
id_para_index = df_limpo.reset_index().set_index('ID')['index']
df_cartoes['id_jogo'] = df_cartoes['partida_id'].map(id_para_index)

#ELIMINANDO COLUNAS DESNECESSARIAS
df_cartoes = df_cartoes.drop(['partida_id','rodata','clube','num_camisa','Pnome','Unome','posicao'], axis=1)
df_cartoes = df_cartoes.rename(columns={'cartao': 'e_tipo', 'minuto': 'n_minuto_cartao'})
```

## Código para inserção de time_participa_campeonato

```python
# CONSTRUINDO INSERÇÕES DA TABELA TIME_PARTICIPA_CAMPEONATO

# Pegando só a data do jogo, time mandante e visitante
df_time_campeonato = pd.read_csv("dataset/campeonato-brasileiro-full.csv")
# Axis=1 indica que é pra remover a coluna
df_time_campeonato = df_time_campeonato.drop(["ID", "rodata", "hora", "formacao_mandante", "formacao_visitante", "tecnico_mandante", "tecnico_visitante", "vencedor", "arena", "mandante_Placar", "visitante_Placar", "mandante_Estado", "visitante_Estado"], axis=1)
display(df_time_campeonato)
```

```python
# Deixando só o ano
df_time_campeonato['ano'] = df_time_campeonato['data'].str[-4:]
df_time_campeonato['ano'] = df_time_campeonato['ano'].astype(int)

# Juntando mandante e visitante só em time
df_time_campeonato = df_time_campeonato.melt(id_vars=['ano'], 
                   value_vars=['mandante', 'visitante'], 
                   value_name='time')
df_time_campeonato = df_time_campeonato.drop(columns=["variable"])

df_time_campeonato = df_time_campeonato.drop_duplicates()

anos_para_remover = range(2003, 2014)
condicao = df_time_campeonato['ano'].isin(anos_para_remover)
df_time_campeonato = df_time_campeonato[~condicao].copy()

display(df_time_campeonato)
```

```python
# Agrupando os times por ano
times_por_ano = df_time_campeonato.groupby('ano')['time'].apply(lambda x: sorted(list(x)))
display(times_por_ano)

numero_de_times_por_ano = times_por_ano.str.len()
display(numero_de_times_por_ano)

# Por algum motivo, o de 2021 esta errado, e tem quatro times a mais, excluindo:
times_de_2021 = times_por_ano.loc[2021]
times_para_remover = ["Botafogo-RJ", "Vasco", "Coritiba", "Goias"]
nova_lista_2021 = [time for time in times_de_2021 if time not in times_para_remover]

times_por_ano.loc[2021] = nova_lista_2021
numero_de_times_por_ano = times_por_ano.str.len()
display(numero_de_times_por_ano)
```

```python
# Criando a consulta

# 1. Preparar o início do comando SQL e uma lista para guardar as linhas de valores
sql_header = """INSERT INTO time_participa_campeonato
(c_nome_time, c_nome_campeonato, d_ano_campeonato)
VALUES\n"""

# Lista para guardar cada linha -> ('Time', 'Camp', Ano)
values_list = []

# 2. Iterar sobre a Series (ano por ano)
# .items() nos dá o índice (ano) e o valor (lista_de_times)
for ano, lista_de_times in times_por_ano.items():
    
    # Para cada ano, iterar sobre a lista de times
    for time in lista_de_times:
        
        # Tratamento de segurança: se um time tiver aspas no nome (ex: O'Higgins),
        # o SQL quebraria. Trocamos ' por '' para evitar isso.
        time_seguro = time.replace("'", "''")
        
        # Monta a string do campeonato
        nome_campeonato = f"Brasileirão {ano}"
        
        # Formata a linha de VALUES no padrão desejado
        # Ex: ('Palmeiras', 'Brasileirão 2023', 2023)
        linha_formatada = f"('{time_seguro}', '{nome_campeonato}', {ano})"
        
        # Adiciona a linha formatada à nossa lista
        values_list.append(linha_formatada)

# 3. Juntar tudo em uma única string de SQL
# Juntamos todas as linhas da lista com ",\n" para que fiquem separadas por vírgula e quebra de linha
sql_values = ",\n".join(values_list)

# Monta o comando final
sql_completo = sql_header + sql_values + ";"

# 4. Salvar o resultado em um arquivo de texto
nome_do_arquivo = 'inserts/insert_time_participa_campeonato.txt'
try:
    with open(nome_do_arquivo, 'w', encoding='utf-8') as f:
        f.write(sql_completo)
    print(f"Arquivo '{nome_do_arquivo}' criado com sucesso!")
    print("Ele contém todos os comandos INSERT prontos para serem usados no seu banco de dados.")

except Exception as e:
    print(f"Ocorreu um erro ao salvar o arquivo: {e}")

```
