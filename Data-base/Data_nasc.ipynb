```python
import pandas as pd

DIR = ["dataset/fm2023.csv", "dataset/fm22data.csv", "dataset/fm21data.csv", "dataset/fm20data.csv"]

lista_de_dataframes = []

for arquivo_csv in DIR:
    try:
        df_temp = pd.read_csv(arquivo_csv, on_bad_lines='skip')
        df_temp.columns = df_temp.columns.str.strip()
        df_temp = df_temp[['Name', 'DoB','Division','Based','Club']]
        lista_de_dataframes.append(df_temp)
    except ValueError as e:
        print(f"Aviso: Colunas 'Name' ou 'DoB' não encontradas em {arquivo_csv}. Pulando arquivo. Detalhe: {e}")
df = pd.concat(lista_de_dataframes, ignore_index=True)

display(df)
```

```python
df_brasil = df[df['Division'].str.contains('Brazilian', case=False, na=False)]
display(df_brasil)
```

```python
import numpy as np
"""
df_brasil['posicao'] = df_brasil['posicao'].str.strip()
df_brasil['posicao'] = df_brasil['posicao'].str.upper()
"""
df_brasil['Name'] = df_brasil['Name'].str.strip()
df_brasil['Name'] = df_brasil['Name'].str.upper()
"""
df_brasil['clube'] = df_brasil['clube'].str.strip()

df_brasil['num_camisa'] = df_brasil['num_camisa'].astype('Int64')
"""
df_brasil['Name'] = df_brasil['Name'].astype(str)

df_brasil['Pnome'] = df_brasil['Name'].str.split().str[0]
df_brasil['Unome'] = df_brasil['Name'].str.split().str[-1]
condicao = df_brasil['Pnome'] == df_brasil['Unome']
df_brasil.loc[condicao, 'Unome'] = np.nan
df_brasil = df_brasil.drop('Name', axis=1)
#df_brasil['d_data_nascimento'] = pd.NA

#colunas = ['Pnome','Unome','num_camisa','d_data_nascimento','posicao','clube']
#df_brasil = df_brasil[colunas].copy()
#df_brasil = df_brasil.rename(columns={'Pnome':'c_Pnome_jogador','Unome':'c_Unome_jogador','num_camisa': 'n_camisa','posicao': 'c_posicao','clube': 'c_nome_time'})


display(df_brasil)

#df_brasil.to_csv("inserts/jogadores_brasileirao.csv", index = False)
```

```python
df_jogadores = pd.read_csv("inserts/arquivos_csvs/jogadores_brasileirao.csv")
display(df_jogadores)
```

```python
# 1. CRIE UMA VERSÃO DE df_brasil SEM NOMES DUPLICADOS
#    Isso garante que cada jogador só aparecerá uma vez.
#    'keep="first"' mantém a primeira ocorrência de cada jogador e remove as subsequentes.
df_brasil_sem_duplicatas = df_brasil.drop_duplicates(subset=['Pnome', 'Unome'], keep='first')

# 2. AGORA, FAÇA O MERGE USANDO O DATAFRAME LIMPO
#    Como agora só existe uma correspondência possível por nome, nenhuma linha será adicionada.
df_atualizado = pd.merge(
    df_jogadores,
    df_brasil_sem_duplicatas[['Pnome', 'Unome', 'DoB']], # Pega apenas as colunas necessárias
    left_on=['c_Pnome_jogador', 'c_Unome_jogador'],
    right_on=['Pnome', 'Unome'],
    how='left'
)

# 3. ATUALIZE A COLUNA DE DATA DE NASCIMENTO E LIMPE
#    Esta parte continua igual, usando .fillna() para preencher apenas os vazios.
df_atualizado['d_data_nascimento'] = df_atualizado['d_data_nascimento'].fillna(df_atualizado['DoB'])

# Remove as colunas auxiliares do merge que não são mais necessárias
df_final = df_atualizado.drop(columns=['Pnome', 'Unome', 'DoB'])

# Opcional: verifique se o número de linhas permaneceu o mesmo
print(f"Linhas em df_jogadores original: {len(df_jogadores)}")
print(f"Linhas em df_final após a atualização: {len(df_final)}")

display(df_final)
```

```python
# Conta quantos valores na coluna 'd_data_nascimento' são nulos
jogadores_sem_data = df_final["d_data_nascimento"].isnull().sum()

# Exibe o resultado de forma clara
print(f"Total de jogadores que continuam sem data de nascimento: {jogadores_sem_data}")
print(f"Total de jogadores com data de nascimento: {len(df_final)-jogadores_sem_data}")
```
